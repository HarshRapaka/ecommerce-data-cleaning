```python
import pandas as pd
df = pd.read_csv("../data/raw_dirty_ecommerce_data.csv")
```

```python
print("Duplicates before: ", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates after: ", df.duplicated().sum())
```

```python
df = df.rename(columns={
    'product_name_lenght': 'product_name_length',
    'product_description_lenght': 'product_description_length'
})
```

```python
df["price"] = df["price"].astype(str).str.extract(r'(\d+\.?\d)')[0]
df["price"] = pd.to_numeric(df["price"], errors='coerce')

#remove negative + invalid
df=df[df["price"]>= 0]
```

```python
print("Invalid prices removed: ", df["price"].isnull().sum())
```

```python
df["order_purchase_timestamp"] = pd.to_datetime(
    df["order_purchase_timestamp"],
    format = "mixed",
    dayfirst=True,
    errors = "coerce"
)
```

```python
print("Invalid dates: ", df["order_purchase_timestamp"].isnull().sum())
```

```python
df['customer_city'] = df['customer_city'].astype(str).str.strip().str.lower()
```

```python
df['customer_city'] = df['customer_city'].str.replace(r'\s+', ' ', regex=True)
```

```python
# Fill Categorical
df["product_category_name"] = df["product_category_name"].fillna("unknown")

# Drop critical missing
df = df.dropna(subset=["price", "order_purchase_timestamp"])
```

```python
print("Remaining Null values:\n", df.isnull().sum())
```

```python
df.to_csv("cleaned_ecommerce_data.csv", index=False)
```

```python
print("\nFINAL CHECKS")

print("Shape:", df.shape)
print("Duplicates:", df.duplicated().sum())
print("Nulls:\n", df.isnull().sum())
print("Data Types:\n", df.dtypes)
```

```python
#assert df.duplicated().sum() == 0
#assert df['price'].min() >= 0
```

```python
df = df.drop_duplicates(subset=[
    'order_id',
    'product_id',
    'order_item_id'
])
```

```python
print("Duplicates after fix: ", df.duplicated().sum())
```

```python
date_cols = [
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'shipping_limit_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='mixed', dayfirst=True, errors='coerce')
```

```python
df = df.dropna(subset=['order_approved_at'])
```

```python
df = df.dropna(subset=[
    'payment_type',
    'payment_value',
    'payment_installments'
])
```

```python
cols_fill = [
    'product_name_length',
    'product_description_length',
    'product_photos_qty'
]

for col in cols_fill:
    df[col] = df[col].fillna(0)
```

```python
cols_phys = [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]

df = df.dropna(subset=cols_phys)
```

```python
int_cols = [
    'order_item_id',
    'payment_sequential',
    'payment_installments'
]

for col in int_cols:
    df.loc[:, col] = df[col].astype(int)
```

```python
print("FINAL SHAPE: ", df.shape)
print("DUPLICATES: ", df.duplicated().sum())
print("NULL:\n", df.isnull().sum())
print("DTYPES:\n", df.shape)
```

```python
df.to_csv("../data/cleaned_ecommerce_data.csv", index=False)
```
